```python
import pandas as pd 
import numpy as np 
import time
import datetime
```

```python
def dtprocess(x):
    y=datetime.datetime.strptime(x, '%d.%m.%Y \n %H:%M:%S')
    return y
```

```python
def dataset(directory,header_index,rename):
    df=pd.read_excel((directory),header=header_index).dropna(axis=1,how='all')
    df=df.rename(columns={'Time':('Time'+rename), 'Value':('Value'+rename)})
    return df
```

```python
OMS103= dataset(r"D:\University\Third Year\Third Year Project\Data Recodings 8.12.2017- 27.02.2020\Room 103_5th Floor_SouthWest\OperationModeState_0103.xlsx",4,'OMS')
OOS103= dataset(r"D:\University\Third Year\Third Year Project\Data Recodings 8.12.2017- 27.02.2020\Room 103_5th Floor_SouthWest\OnOffState_0103.xlsx",4,'OOS')
RT103=dataset(r"D:\University\Third Year\Third Year Project\Data Recodings 8.12.2017- 27.02.2020\Room 103_5th Floor_SouthWest\RoomTemp_0103.xlsx",4,'RT')
ST103=dataset(r"D:\University\Third Year\Third Year Project\Data Recodings 8.12.2017- 27.02.2020\Room 103_5th Floor_SouthWest\SetTemp_0103.xlsx",4,'ST')
```

```python

def merge_cols(cols):
    #for df in cols:
       # df=df.where(((df.iloc[:,2]).diff())<=630)
    merged_ds=pd.concat(cols,axis=1).dropna(axis=0,how='any')
    return(merged_ds)
ds=[OMS103,OOS103,RT103,ST103]
merged_ds=merge_cols(ds)
merged_ds=merged_ds[:]
print(merged_ds.head())

len(merged_ds)
```

```python
times=[df.iloc[:,i:i+2].dropna().values for i in range(0,8,2)]
print(times[0][3])
print(len(times))


for col in times:
    for row in col:
        row[0]=datetime.datetime.strptime(row[0], '%d.%m.%Y \n %H:%M:%S')
rows=[]

for time_0 in times[0]:
    row=[time_0[0],time_0[1]]
    valid_row=True
    for j in range(1,(len(times))):
        if len(times[j])==0:
            valid_row=False
        if valid_row:
            for i in range(len(times[j])):
                if times[j][i][0]>row[0]:
                    if ((times[j][i][0]-row[0]).seconds)<630:
                        row.extend(times[j][i])
                        times[j]=times[j][i+1:]
                        break

                    else:
                        valid_row=False
                        times[j]=times[j][i:]
                        break
                        
    if valid_row:
        rows.append(row)

indoor_ds= pd.DataFrame(rows, columns=df.columns[:8])
```
